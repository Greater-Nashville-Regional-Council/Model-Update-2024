```python
import pandas as pd
import sqlite3 as sq
import matplotlib as mpl
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
```

```python
conn = sq.connect('../../Data-Pipelines/Outputs/Dem_Transpo_Housing_Collection.db')
sql_query = pd.read_sql('SELECT * FROM [WPURBANSIM_PopulationHousehold_CurrentandProjected_Annual_WideFormat_Noreplacement]', conn)
initial = pd.DataFrame(sql_query)
initial = initial[['NAME', 'Population 2017', 'Population 2020', 'Population 2025', 'Population 2035', 'Population 2045']]
thelist = ['Cheatham County, Tennessee', 'Davidson County, Tennessee', 'Dickson County, Tennessee', 'Houston County, Tennessee', 
           'Humphreys County, Tennessee', 'Maury County, Tennessee', 'Montgomery County, Tennessee', 'Robertson County, Tennessee', 
           'Rutherford County, Tennessee', 'Stewart County, Tennessee', 'Sumner County, Tennessee', 'Williamson County, Tennessee', 
           'Wilson County, Tennessee', 'Trousdale County, Tennessee', 'GNRC', 'MPO', 'GNRC Region']
initial = initial.loc[initial['NAME'].isin(thelist)].reset_index(drop = True)
initial = initial.rename(columns = {'Population 2017': '2017', 'Population 2020': '2020', 'Population 2025': '2025', 
                                      'Population 2035': '2035', 'Population 2045': '2045'})

#first option
#initial['Base'] = '2023'
# second option
initial = initial.melt(id_vars = ['NAME'], var_name = 'Year', value_name = 'Population 2017 Base')
```

```python
conn = sq.connect('../../Data-Pipelines/Outputs/Dem_Transpo_Housing_Collection.db')
sql_query = pd.read_sql('SELECT * FROM [WPURBANSIM_PopulationHousehold_CurrentandProjected_Annual_WideFormat_Noreplacement_2023Base]', conn)
initial1 = pd.DataFrame(sql_query)
initial1 = initial1[['NAME', 'Population 2017', 'Population 2020', 'Population 2025', 'Population 2035', 'Population 2045']]
thelist = ['Cheatham County, Tennessee', 'Davidson County, Tennessee', 'Dickson County, Tennessee', 'Houston County, Tennessee', 
           'Humphreys County, Tennessee', 'Maury County, Tennessee', 'Montgomery County, Tennessee', 'Robertson County, Tennessee', 
           'Rutherford County, Tennessee', 'Stewart County, Tennessee', 'Sumner County, Tennessee', 'Williamson County, Tennessee', 
           'Wilson County, Tennessee', 'Trousdale County, Tennessee', 'GNRC', 'MPO', 'GNRC Region']
initial1 = initial1.loc[initial1['NAME'].isin(thelist)].reset_index(drop = True)
initial1 = initial1.rename(columns = {'Population 2017': '2017', 'Population 2020': '2020', 'Population 2025': '2025', 
                                      'Population 2035': '2035', 'Population 2045': '2045'})

#first option
#initial1['Base'] = '2023'
# second option
initial1 = initial1.melt(id_vars = ['NAME'], var_name = 'Year', value_name = 'Population 2023 Base')
```

```python
conn = sq.connect('../../Data-Pipelines/Outputs/Dem_Transpo_Housing_Collection.db')
sql_query = pd.read_sql('SELECT * FROM [TNSDC_PopProjections2020Base]', conn)
other = pd.DataFrame(sql_query)
thelist = ['Cheatham County, Tennessee', 'Davidson County, Tennessee', 'Dickson County, Tennessee', 'Houston County, Tennessee', 
           'Humphreys County, Tennessee', 'Maury County, Tennessee', 'Montgomery County, Tennessee', 'Robertson County, Tennessee', 
           'Rutherford County, Tennessee', 'Stewart County, Tennessee', 'Sumner County, Tennessee', 'Williamson County, Tennessee', 
           'Wilson County, Tennessee', 'Trousdale County, Tennessee', 'GNRC', 'MPO', 'GNRC Region']
other = other.loc[other['NAME'].isin(thelist)]
other = other[['NAME', '2020', '2025', '2035', '2045']]

# second option
other = other.melt(id_vars = ['NAME'], var_name = 'Year', value_name = 'Population TNSDC')
```

```python
other.head()
```

```python
data = initial.merge(initial1, on = ['NAME', 'Year'], how = 'outer')
data = data.merge(other, on = ['NAME', 'Year'], how = 'outer')
```

```python
data['Difference 2023-2017 Base'] = data['Population 2023 Base'] - data['Population 2017 Base']
data['Difference % 2023-2017 Base'] = ((data['Population 2023 Base'] - data['Population 2017 Base'])/data['Population 2017 Base'])*100

data['Difference 2023-TNSDC Base'] = data['Population 2023 Base'] - data['Population TNSDC']
data['Difference % 2023-TNSDC Base'] = ((data['Population 2023 Base'] - data['Population TNSDC'])/data['Population TNSDC'])*100
```
