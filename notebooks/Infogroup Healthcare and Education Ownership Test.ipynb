This notebook examines the ownership column to see if it is quality, useable data from DataAxel for NAICS industries where there is typically a mix of private and public ownership. The concensus is no, it is not useable. Keeping this file for reference.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 150)
```

```python
data = pd.read_csv('../data/infogroupfull.csv')
```

```python
data['naics_code'] = data['naics_code'].astype(str)
data['2dignaics'] = data['naics_code'].str.slice(start = 0, stop = 2)
```

```python
data = data.loc[data['2dignaics'] == '62']
```

```python
data['OWNERSHIP DESCRIPTION'] = data['OWNERSHIP DESCRIPTION'].fillna('Empty Field')
```

```python
thing = list(data['OWNERSHIP DESCRIPTION'].unique())
```

```python
thing
```

```python
for description in thing:
    temp_df = data.copy()
    temp_df = temp_df.loc[temp_df['OWNERSHIP DESCRIPTION'] == '{}'.format(description)]
    print('{}'.format(description) + " has this many items: " + str(len(temp_df)))
```

I identified many of these as potentially public ownership. The following code indexes into the identified fields.

```python
temp = data.loc[data['OWNERSHIP DESCRIPTION'] == 'Empty Field']
temp.head(10)
```

```python
temp = data.loc[data['OWNERSHIP DESCRIPTION'] == 'RADIOLOGY']
temp.head(10)
```

So for example Nashville General Hospital shows up in radiology and in "GOVERNMENT OWNED"...
